```python
from numpy import random, sqrt, log, sin, cos, pi
from pylab import show,hist,subplot,figure
from matplotlib import pyplot
import numpy
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
U1 = numpy.random.random_sample(10000)
print(U1)

U2 = numpy.random.random_sample(10000)
print(U2)
```

```python
figure()
subplot(221) # the first row of graphs
hist(U1,bins=numpy.arange(0.,1.01, 0.01),normed=False);     # contains the histograms of u1 and u2 
subplot(222)
hist(U2,bins=numpy.arange(0.,1.01, 0.01),normed=False);

```

```python
Z0 = sqrt(-2*log(U1))*cos(2*pi*U2)

Z1 = sqrt(-2*log(U1))*sin(2*pi*U2)

figure()
subplot(221)
binwidth=0.1
bins=numpy.arange(-5.,5.1, binwidth)
pyplot.hist(Z0,bins,normed='true',color='blue');
subplot(222)
binwidth=0.1
bins=numpy.arange(-5.,5.1, binwidth)
pyplot.hist(Z1,bins,normed='true',color='blue');
```
